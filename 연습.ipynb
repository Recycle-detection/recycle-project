```python
def modelload(weightdir):
    import os
    from ultralytics import YOLO
    path=os.path.join(weightdir+'/ultralytics/runs/detect/train2/weights/',"best.pt")
    model=YOLO(path)
    return weightdir
```

```python
a=modelload(weightdir='/Users/i/Desktop/object-detection')
a
```

```python
def modelpredict(imgdir,imgname):
    import os 
    import cv2
    from ultralytics import YOLO
    import numpy as np
    from PIL import Image
    import matplotlib.pyplot as plt
    path1=os.path.join(imgdir+'/ultralytics/ultralytics/yolo/data/data/test/images/',imgname)
    im1=cv2.imread(path1)
    model.predict(source=im1, save=True, save_txt=True) 

    #시각화
    %matplotlib inline
    path=os.path.join(imgdir+'/ultralytics/runs/detect/predict9/',"image0.jpg")
    image = Image.open(path)
    img = np.array(image)
    plt.figure(figsize=(30,30))
    plt.imshow(img)
    plt.show()
```
